<span style="font-size: 14pt; font-weight: bold; color: #0098cd;">¿Cuántas palabras totales hay en los comentarios del corpus?</span>

````python
# Incluye aquí el código generado para poder responder a tu pregunta
total_words = 0

for text in data["CONTENIDO A ANALIZAR"].fillna(''):
    doc = nlp(text)
    palabras = [token.text for token in doc if not token.is_punct and not token.is_space]
    total_words += len(palabras)

print(total_words)
````

Se cargaron los primeros 100,000 registros del archivo CSV utilizando la librería *pandas*. Además, se especificó el encoding `latin-1` para evitar errores de codificación:

### 🔹 Paso 4: Procesamiento de texto con NLP

Se recorrió cada comentario utilizando un bucle `for`. Para cada texto:

1. Se procesa con el modelo de lenguaje cargado previamente (**nlp**).

2. Se tokeniza el texto (es decir, se divide en palabras o unidades léxicas).

3. Se filtran los tokens para eliminar:

   * signos de puntuación (`token.is_punct`)
   * espacios (`token.is_space`)

4. Se cuentan únicamente las palabras válidas.

```python id="o3k7gx"
for text in data_total["CONTENIDO A ANALIZAR"].fillna(''):
    doc = nlp(text)
    palabras = [token.text for token in doc if not token.is_punct and not token.is_space]
    total_words += len(palabras)
```

Este enfoque permite obtener un conteo más preciso de palabras, evitando considerar caracteres irrelevantes.

---

### 🔹 Paso 5: Resultado final

Finalmente, se imprime el total de palabras acumuladas en la variable **total_words**:

```python id="pry3i4"

```
