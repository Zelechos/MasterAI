<span style="font-size: 14pt; font-weight: bold; color: #0098cd;">Pregunta 4.</span>
<span style="font-size: 14pt; font-weight: bold; color: #0098cd;">Considerando dos grupos de comentarios (odio y no odio) ¿Cuál el número promedio de palabras en los comentarios de cada grupo?</span>

````python
# Filtrar solo comentarios
data_filtrado = data[data["TIPO DE MENSAJE"] == "COMENTARIO"]

# Limpiar texto
data["CONTENIDO A ANALIZAR"] = data_filtrado["CONTENIDO A ANALIZAR"].fillna('')

# Tokenización eficiente con pipe
docs = list(nlp.pipe(data["CONTENIDO A ANALIZAR"].astype(str)))

# Conteo de palabras (filtrando puntuación y espacios)
data["num_words"] = [
    len([t for t in doc if not t.is_punct and not t.is_space])
    for doc in docs
]

# Clasificación de comentarios en base a la intensidad
def classify(intensity):
    if pd.isna(intensity) or intensity == 0:
        return "no_odio"
    else:
        return "odio"

data["grupo"] = data["INTENSIDAD"].apply(classify)

# Cálculo de promedio por grupo
response = (
    data.groupby("grupo")["num_words"]
    .mean()
    .reindex(["odio", "no_odio"], fill_value=0)
)

print(response)
````

````python
# Filtrar solo comentarios
data_filtrado = data[data["TIPO DE MENSAJE"] == "COMENTARIO"]

# Limpiar texto
data["CONTENIDO A ANALIZAR"] = data_filtrado["CONTENIDO A ANALIZAR"].fillna('')

# Procesar textos en batch (eficiente)
docs = list(nlp.pipe(data["CONTENIDO A ANALIZAR"].astype(str)))

# Conteo de oraciones usando spaCy
data["num_sentences"] = [len(list(doc.sents)) for doc in docs]

# Clasificación mejorada (igual que antes)
def classify(intensity):
    if pd.isna(intensity) or intensity == 0:
        return "no_odio"
    else:
        return "odio"

data["grupo"] = data["INTENSIDAD"].apply(classify)

# Promedio de oraciones por grupo
resultado = (
    data.groupby("grupo")["num_sentences"]
    .mean()
    .reindex(["odio", "no_odio"], fill_value=0)
)

print(resultado)
````

### **Comparación de medias + test estadístico (t-test)**

````python
from scipy.stats import ttest_ind

def lexical_diversity(text):
    palabras = text.split()
    if len(palabras) == 0:
        return 0
    return len(set(palabras)) / len(palabras)

# Filtrar solo comentarios
data_filtrado = data[data["TIPO DE MENSAJE"] == "COMENTARIO"]

data["lexical_diversity"] = data_filtrado["CONTENIDO A ANALIZAR"].fillna('').apply(lexical_diversity)

features = [
    "num_words",
    "num_sentences",
    "has_ner",
    "avg_word_length",
    "upper_ratio",
    "punctuation_count",
    "lexical_diversity"
]

odio = data[data["grupo"] == "odio"]
no_odio = data[data["grupo"] == "no_odio"]

results = []

# Agregamos evaluacion por Ttest
for f in features:
    t_stat, p_value = ttest_ind(
        odio[f].dropna(),
        no_odio[f].dropna(),
        equal_var=False
    )

    results.append({
        "feature": f,
        "mean_odio": odio[f].mean(),
        "mean_no_odio": no_odio[f].mean(),
        "p_value": p_value
    })

stats_df = pd.DataFrame(results)

print(stats_df.sort_values("p_value"))
````

### **Interpretación de Cohen’d**

````python
import numpy as np

def cohens_d(x, y):
    nx, ny = len(x), len(y)
    dof = nx + ny - 2
    pooled_std = np.sqrt(((nx - 1)*x.std()**2 + (ny - 1)*y.std()**2) / dof)
    return (x.mean() - y.mean()) / pooled_std

stats_df["cohens_d"] = stats_df["feature"].apply(
    lambda f: cohens_d(odio[f].dropna(), no_odio[f].dropna())
)

print(stats_df.sort_values("cohens_d", ascending=False))
````
